```python
import warnings
warnings.filterwarnings('ignore')

import os
import sys
from pathlib import Path

import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
p = '/vol/ml/apartin/projects/cell-line-drug-sensitivity/models/candle_challenge_prb/cnt/hrd/summary.csv'
df = pd.read_csv(p)
df
```

```python
plt.plot(df['weps'], df['ceps'], '-or')
plt.xlabel('weps')
plt.ylabel('ceps')
plt.grid(True)
```

```python
plt.plot(df['weps'], df['runtime_sec'], '-ob')
plt.xlabel('weps')
plt.ylabel('runtime')
plt.grid(True)
```

```python
fig, ax1 = plt.subplots()

ax1.plot(df['weps'], df['ceps'], '-om')
ax1.set_xlabel('weps')
ax1.set_ylabel('ceps', color='m')
ax1.tick_params('y', colors='m')
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(df['weps'], df['runtime_sec'], '-ob')
ax2.set_ylabel('runtime', color='b')
ax2.tick_params('y', colors='b')

fig.tight_layout()
plt.show()
```
